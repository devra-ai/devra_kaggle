# Marvel Movies Box Office Analysis
This notebook provides an analysis of the Marvel Movies dataset, focusing on various metrics such as worldwide gross, budget, and scores.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Marvel Movies.csv'
marvel_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
marvel_df.head()
```

## Pie Chart of Movies by Worldwide Gross Revenue

```python
# Create a pie chart of movies by worldwide gross revenue
revenue_data = marvel_df.groupby('movie')['worldwide gross ($m)'].sum()
plt.figure(figsize=(10, 8))
revenue_data.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Marvel Movies by Worldwide Gross Revenue')
plt.ylabel('')
plt.show()
```
